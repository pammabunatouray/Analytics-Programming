```python
#Grading ID: KP1135
#The purpose of this program is to create an interative Python script that allows a meteorologist to enter recorded wind mileages for tornadoes, then calculates each tornado's mean wind gust and category, and displays a bar chart of the categories assigned (0-5). Divide your application's work between two functions named input_wind_gusts and summarize_tornadoes. The first function will prompt the user for how many tornadoes and how many gust measurements for each tornado, then use loops to input this data into a 2-d list of tornado wind gusts. The function will return this 2-d list of tornadoes.
#Due: Tuesday,March 21
#Class: CIS 443-01
import matplotlib.pyplot as plt
import seaborn as sns

def input_wind_gusts():
    num_tornadoes = int(input("Enter the number of tornadoes: "))
    tornadoes = []
    for i in range(num_tornadoes):
        num_gusts = int(input(f"Enter the number of wind gust measurements for tornado {i+1}: "))
        gusts = []
        for i in range(num_gusts):
            gust = float(input(f"Enter the wind gust measurement {i+1} for tornado {i+1}: "))
            gusts.append(gust)
        tornadoes.append(gusts)
    return tornadoes

def summarize_tornadoes(tornadoes):
    means = []
    for gusts in tornadoes:
        mean_gust = sum(gusts)/ len(gusts)
        if 65 <= mean_gust <= 85:
            category = 0
        elif 86 <= mean_gust <= 110:
            category = 1
        elif 111 <= mean_gust <= 135:
            category = 2
        elif 136 <= mean_gust <= 165:
            category = 3
        elif 166 <= mean_gust < 200:
            category = 4
        else:
            category = 5
        means.append((mean_gust, category))
    
    categories = [x[1] for x in means]
    category_counts = [categories.count(x) for x in range(6)]
    category_labels = ["0", "1", "2", "3", "4", "5"]
    
    sns.set_style("whitegrid")
    plt.bar(category_labels, category_counts)
    plt.title("Tornado Category")
    plt.xlabel("Category")
    plt.ylabel("Frequency")
    plt.show()
tornadoes = input_wind_gusts()
summarize_tornadoes(tornadoes)
```
